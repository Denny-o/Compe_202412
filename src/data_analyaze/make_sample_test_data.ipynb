```python
def calculate_missing_percentage(df):
    """ データフレームの各カラムの欠損率を計算する関数 
    Parameters: df (pd.DataFrame): 欠損率を計算する対象のデータフレーム 
    Returns: pd.Series: 各カラムの欠損率を含むシリーズ 
    """ 
    missing_percentage = df.isnull().mean() * 100 
    return missing_percentage
```

```python
import pandas as pd
import numpy as np

def calculate_column_averages_with_nan(df):
    """
    データフレームの各列の平均値を計算し、数値型以外の列にはNaNを挿入する関数
    
    Parameters:
        df (pd.DataFrame): 平均値を計算する対象のデータフレーム
    
    Returns:
        pd.Series: 各列の平均値を含むシリーズ。ただし、数値型以外の列にはNaNを含む
    """
    averages = {}
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            averages[column] = df[column].mean()
        else:
            averages[column] = np.nan
    return pd.Series(averages)
```
